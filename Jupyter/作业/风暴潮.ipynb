```python
import pandas as pd
import matplotlib.pylab as plt
import datetime as dt
```

```python
with open(r"E:\Data\风暴潮.txt",encoding="gbk") as f:
    t = f.readlines()
    range_ = [i.split()[0] for i in t]
    levle = [i.split()[1] for i in t]
    t_name = [i.split()[2] for i in t]
    time_ = [i.split()[3] for i in t]
```

```python
data = pd.DataFrame({'range':range_,'levle':levle,'t_name':t_name,'time_':time_})
data
```

```python
data['time_'] = pd.to_datetime(data['time_'],format='mixed')
data.sort_values(by='time_',ascending=True,inplace=True)
data
```

```python
plt.plot(data['time_'],data['levle'])
```
