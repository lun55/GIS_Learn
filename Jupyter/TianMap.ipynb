```python
def getlocation(application_key,searchWord):
    header = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
        'Cookie': 'HWWAFSESID=701f9b2a720ea2f8fe; HWWAFSESTIME=1701775847088; TDTSESID=7bdaea5fcf680600154bf7cdcfa8cf64|8a99c1bdeb67a53f52230a32ac84b973'
        }
    url =  'http://api.tianditu.gov.cn/geocoder?ds={"keyWord":"'+searchWord+'"}&tk='+application_key
    try:
        response = requests.get(url,headers=header)
        if response.status_code==200:
            adress_json = response.json()
            lon = adress_json['location']['lon']
            lat = adress_json['location']['lat']
            print([lon,lat])
            return [lon,lat]
    except:
        return 0
```
